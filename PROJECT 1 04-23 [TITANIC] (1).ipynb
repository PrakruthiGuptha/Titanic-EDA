```python
class_mean_ages = df.groupby(['Gender','Pclass'])['Age'].mean()
```

```python
class_mean_ages
```

Here we are finding the mean of age, of male and female in each class which is actually helping us to replace, and get much accurate values.

```python
def replace_age(row):
    if pd.isnull(row['Age']):
        return class_mean_ages[row['Gender'],row['Pclass']]
    else:
        return row['Age']

df.Age =  df.apply(replace_age,axis= 1)
df.info()
```

In the above attribute we are replacing the null values with the mean of age. And we have joined another attribute saying df.info(), so it will first will the null values and then give information. As we can see there are no null values in age column.

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```
